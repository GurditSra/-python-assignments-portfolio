```python
import numpy as np
import matplotlib.pyplot as plt

# heat capacity parameters
# for CV_O2 and CV_Prop, which is propane  <-- you will need CV in for the assignment
# here we use a logistic function model for heat capacity
# CV(T) = a1+b1/(1+exp((-T+c1)/d1))
a_O = 20.6
b_O = 4.6
c_O = 5.19
d_O = 102

a_P = -69.8
b_P = 278.2
c_P = 313
d_P = 390

Tmin = 300
Tmax = 730
dT = 10
N = int((Tmax-Tmin)/dT)+2
T = np.zeros(N) # build and array of N spots
T_ = Tmin
for i in range(0,N):
    T[i] = T_
    T_ += dT

Vin = 0.25*np.pi*(0.07**2)
R = 8.314  # for pascals, which is for joules
r = 0.08314   # for bar
delta_Sv = np.zeros(N)

CV_O2 = np.zeros(N)
CV_prop = np.zeros(N)

for i in range(0,N):
    CV_O2[i] = a_O+b_O/(1+np.exp((-T[i]+c_O)/d_O))
    CV_prop[i] = a_P+b_P/(1+np.exp((-T[i]+c_P)/d_P))

n_tot = (Vin)/(0.08314*298.15)
n_P = n_tot/6
n_O = n_tot*5/6
P = np.zeros(N)
V = np.zeros(N)

for i in range(0,N-1):
    CV_T = CV_O2[i]*n_O + CV_prop[i]*n_P
    FofT = CV_T/T[i]
    CV_T1 = CV_O2[i+1]*n_O + CV_prop[i+1]*n_P
    FofT1 = CV_T1/T[i+1]
    delta_Sv[i] = ((FofT+FofT1)/2)*dT

    V[i] = Vin*np.exp(-np.sum(delta_Sv)/(n_tot*R))
    P[i] = (n_tot*r*T[i])/V[i]

"""for i in range(0,N):
    CV_T = CV_O2[i-1]*n_O + CV_prop[i-1]*n_P
    FofT = CV_T/T[i-1]
    CV_T1 = CV_O2[i]*n_O + CV_prop[i]*n_P
    FofT1 = CV_T1/T[i]
    delta_Sv[i] = delta_Sv[i-1]+((FofT+FofT1)/2)*dT

    V[i] = Vin*np.exp(-delta_Sv[i]/(n_tot*R))
    P[i] = (n_tot*R*T[i])/V[i]"""

P = P[:-1]
V = V[:-1]


plt.plot(V,P)
plt.xlabel('Volume (L)')
plt.ylabel('Pressure (bar)')
plt.title('Path for Reversible Adiabatic Compression of the Gas to Ignition')
plt.plot(V,P, color='green')
#plt.ylim(0, 140)
#plt.xlim(0, 0.014)
plt.show()
print('Pressure:', P[-1],"", 'Volume:', V[-1], "", 'Temperature:', T[N-1], "", 'Moles:', n_tot, "", 'Compression Ratio:', Vin/V[-1])
```

```python
n_rxn = n_P

#initial temp is flame temp, final temp is that when the piston is at the original volume again
Tinitial = 6160
Tfinal = 3370


# heat capacity parameters for carbon dioxide
a3 = -20
b3 = 70
c3 = 50
d3 = 550

# heat capacity parameters for water
a4 = 20.4
b4 = 39.6
c4 = 2.6
d4 = 932


dT   = -10
N3    = int((Tfinal - Tinitial)/dT) + 1
T3    = np.zeros(N3)
T_   = Tinitial

# creates an array of your temperature range from 6160 K to 3370 K with increments of -10 K
for i in range(0,N3):
    T3[i] = T_
    T_ += dT

#heat capacity calculations
CV_CO2_3 = np.zeros(N3)
CV_H2O_3 = np.zeros(N3)

for i in range(0,N3):
    CV_CO2_3[i] = (a3+(b3/(1+np.exp((-T3[i]+c3)/d3))))
    CV_H2O_3[i] = (a4+(b4/(1+np.exp((-T3[i]+c4)/d4))))

CV_prod2 = 3*n_rxn*CV_CO2_3 + 4*n_rxn*CV_H2O_3

n_prod = 3*n_rxn + 4*n_rxn


dS_V2 = np.zeros(N3)
V2 = np.zeros(N3)
P2 = np.zeros(N3)
dS_V2[0] = 0
V_in2 = V[43]

# range is not inclusive so we add the last entry [-1] at the beginning
# here we break down the integral into a summation (the trapezoidal rule)

for i in range(0,N3):
    if np.all(V2 <= 0.00385):
        dS_V2[i] = dS_V2[i-1] + (CV_prod2[i-1]/T3[i-1] + CV_prod2[i]/T3[i]) * dT * 0.5
        V2[i] = V_in2*np.exp(-dS_V2[i]/(n_prod*R))
        P2[i] = (n_prod*r*T3[i])/V2[i]
    else:
        break #the if condition inside the loop breaks it when the volume exceeds the original volume of the piston

plt.figure(2)
plt.ylabel('Pressure (bar)')
plt.xlabel('Volume (L)')
plt.title('Path for Reversible Adiabatic Expansion of the Gas to Initial Volume')
plt.plot(V2,P2, color='maroon')
print('Pressure:', P2[0],"", 'Volume:', V2[0])
```
